```python
import numpy as np

# Define phases based on 'over' number
def get_phase(over):
    if 1 <= over <= 6:
        return 'Powerplay'
    elif 7 <= over <= 15:
        return 'Middle Overs'
    else:
        return 'Death Overs'

# Apply phase classification
df_ipl['phase'] = df_ipl['over'].apply(get_phase)

# Group by batting team and phase
phase_runrate = df_ipl.groupby(['batting_team', 'phase']).agg(
    total_runs=('total_runs', 'sum'),
    total_overs=('over', lambda x: x.nunique())
).reset_index()

# Calculate run rate = runs / overs
phase_runrate['run_rate'] = (phase_runrate['total_runs'] / phase_runrate['total_overs']).round()

print(phase_runrate.sort_values(by=['batting_team', 'phase']).reset_index(drop=True))
```

```python
phase_runrate_pivot = phase_runrate.pivot(
    index='batting_team', columns='phase', values='run_rate'
).reset_index().fillna(0)

print(phase_runrate_pivot)
```

```python
# ============================================================
#  IPL Comprehensive Bowling Insights (Team-wise)
# ============================================================

import pandas as pd

# ------------------------------------------------------------
# 1 TOTAL RUNS CONCEDED BY EACH TEAM
# ------------------------------------------------------------
runs_conceded = (
    df_ipl.groupby("bowling_team")["total_runs"]
    .sum()
    .sort_values(ascending=False)
    .reset_index(name="Runs Conceded")
)
print(" Total Runs Conceded by Each Team")
print(runs_conceded.to_string(index=False))
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 2 TOTAL WICKETS TAKEN BY EACH TEAM
# ------------------------------------------------------------
# Filter only legitimate dismissals (ignore runouts/extras)
valid_dismissals = [
    "caught", "bowled", "lbw", "stumped",
    "caught and bowled", "hit wicket"
]

wickets_taken = (
    df_ipl[df_ipl["dismissal_kind"].isin(valid_dismissals)]
    .groupby("bowling_team")["player_dismissed"]
    .count()
    .sort_values(ascending=False)
    .reset_index(name="Wickets Taken")
)
print(" Total Wickets Taken by Each Team")
print(wickets_taken.to_string(index=False))
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 3 BOWLING ECONOMY RATE = RUNS CONCEDED / OVERS BOWLED
# ------------------------------------------------------------
overs_bowled = (
    df_ipl.groupby(["bowling_team", "match_id"])["over"]
    .nunique()
    .groupby("bowling_team")
    .sum()
    .reset_index(name="Overs Bowled")
)

economy_df = pd.merge(runs_conceded, overs_bowled, on="bowling_team", how="left")
economy_df["Economy Rate"] = (economy_df["Runs Conceded"] / economy_df["Overs Bowled"]).round(2)
print(" Bowling Economy Rate (Runs per Over)")
print(economy_df.sort_values(by="Economy Rate").to_string(index=False))
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 4 STRIKE RATE = BALLS PER WICKET
# ------------------------------------------------------------
balls_bowled = (
    df_ipl.groupby("bowling_team")["ball"]
    .count()
    .reset_index(name="Balls Bowled")
)

strike_rate_df = pd.merge(balls_bowled, wickets_taken, on="bowling_team", how="left")
strike_rate_df["Strike Rate"] = (
    strike_rate_df["Balls Bowled"] / strike_rate_df["Wickets Taken"]
).round(2)
print(" Bowling Strike Rate (Balls per Wicket)")
print(strike_rate_df.sort_values(by="Strike Rate").to_string(index=False))
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 5 DOT BALL PERCENTAGE = DOT BALLS / TOTAL BALLS  100
# ------------------------------------------------------------
dot_balls = (
    df_ipl[df_ipl["total_runs"] == 0]
    .groupby("bowling_team")["ball"]
    .count()
    .reset_index(name="Dot Balls")
)

dot_ball_df = pd.merge(balls_bowled, dot_balls, on="bowling_team", how="left").fillna(0)
dot_ball_df["Dot Ball %"] = (
    dot_ball_df["Dot Balls"] / dot_ball_df["Balls Bowled"] * 100
).round(2)
print(" Dot Ball Percentage per Team")
print(dot_ball_df.sort_values(by="Dot Ball %", ascending=False).to_string(index=False))
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 6 MOST DISCIPLINED TEAM (LEAST EXTRAS PER OVER)
# ------------------------------------------------------------
extras = (
    df_ipl.groupby("bowling_team")["extra_runs"]
    .sum()
    .reset_index(name="Total Extras")
)

discipline_df = pd.merge(extras, overs_bowled, on="bowling_team", how="left")
discipline_df["Extras per Over"] = (
    discipline_df["Total Extras"] / discipline_df["Overs Bowled"]
).round(2)

most_disciplined = discipline_df.sort_values(by="Extras per Over").head(1)
print(" Most Disciplined Team (Least Extras per Over)")
print(discipline_df.sort_values(by="Extras per Over").to_string(index=False))
print("\n Most Disciplined Team:", most_disciplined["bowling_team"].values[0])
print("\n" + "-"*80 + "\n")

# ------------------------------------------------------------
# 7 DEATH-OVER EFFICIENCY (FEWEST RUNS CONCEDED IN OVERS 1620)
# ------------------------------------------------------------
death_overs = df_ipl[df_ipl["over"].between(16, 20)]
death_efficiency = (
    death_overs.groupby("bowling_team")["total_runs"]
    .sum()
    .sort_values()
    .reset_index(name="Runs Conceded (Death Overs)")
)
print(" Death-Over Efficiency (Fewest Runs Conceded in Overs 1620)")
print(death_efficiency.to_string(index=False))
print("\n Best Death-Over Bowling Team:",
      death_efficiency.head(1)["bowling_team"].values[0])
print("\n" + "="*80)
```

```python
# ============================================================
# Advanced IPL Analytics: Phase/Over, Match-Level, Stats & Extras
# ============================================================
import pandas as pd
import numpy as np

# Assumes `df_ipl` is available in the environment
# Defensive: abort with friendly message if not present
try:
    df_ipl
except NameError:
    raise RuntimeError("DataFrame `df_ipl` not found. Load your dataset into variable `df_ipl` and re-run.")

# Helper: safe column existence check
def has_cols(df, cols):
    return all(c in df.columns for c in cols)

# ---------------------------
# 4. Phase & Over-Level Insights
# ---------------------------
print("\n" + "="*80)
print("4. PHASE & OVER-LEVEL INSIGHTS")
print("="*80 + "\n")

# 4.1 Average runs per over (120)
avg_runs_per_over = df_ipl.groupby('over')['total_runs'].mean().reset_index().rename(columns={'total_runs':'avg_runs'})
print("Average runs per over (overs 120):")
print(avg_runs_per_over.to_string(index=False))
print()

# 4.2 Most productive overs (eg. overs with avg >= threshold)
threshold = 10.0  # example threshold; adjust as needed
productive_overs = avg_runs_per_over[avg_runs_per_over['avg_runs'] >= threshold].sort_values(by='avg_runs', ascending=False)
print(f"Most productive overs (avg runs >= {threshold}):")
print(productive_overs.to_string(index=False))
print()

# 4.3 Most economical overs (lowest avg runs)  example for overs 79 or overall
economical_window = avg_runs_per_over[avg_runs_per_over['over'].between(7,9)]
print("Average runs for overs 79 (economical window):")
print(economical_window.to_string(index=False))
print()

# 4.4 Wickets per over trend
if 'is_wicket' in df_ipl.columns:
    wickets_per_over = df_ipl.groupby('over')['is_wicket'].sum().reset_index().rename(columns={'is_wicket':'wickets'})
    print("Wickets per over trend:")
    print(wickets_per_over.to_string(index=False))
    print()
else:
    print(" Column 'is_wicket' not found  cannot compute wickets per over.\n")

# 4.5 Team performance in different phases: Powerplay (16), Middle (715), Death (1620)
def phase_from_over(o):
    if 1 <= o <= 6:
        return 'Powerplay'
    elif 7 <= o <= 15:
        return 'Middle'
    elif 16 <= o <= 20:
        return 'Death'
    else:
        return 'Other'

df_ipl['phase'] = df_ipl['over'].apply(phase_from_over)

phase_team = df_ipl.groupby(['batting_team','phase']).agg(
    runs=('total_runs','sum'),
    balls=('ball','count')
).reset_index()
phase_team['overs'] = (phase_team['balls']/6).round(2)
phase_team['run_rate'] = (phase_team['runs'] / phase_team['overs']).replace([np.inf, -np.inf], 0).round(2)
print("Team run rate by phase (sample):")
print(phase_team.pivot(index='batting_team', columns='phase', values='run_rate').fillna(0).head(12).to_string())
print()

# 4.6 Compare teams scoring trends across phases (pivot)
phase_pivot = phase_team.pivot(index='batting_team', columns='phase', values='run_rate').fillna(0).reset_index()
print("Phase-wise run rate pivot (Powerplay / Middle / Death):")
print(phase_pivot.to_string(index=False))
print()

# ---------------------------
# 5. Match-Level / Head-to-Head Insights
# ---------------------------
print("\n" + "="*80)
print("5. MATCH-LEVEL & HEAD-TO-HEAD INSIGHTS")
print("="*80 + "\n")

# Prepare a match-level aggregation (one row per match per team)
# We need, per match and batting_team: total runs by that team in the match
team_match_runs = (
    df_ipl.groupby(['match_id','batting_team'])
    .agg(runs_in_match=('total_runs','sum'))
    .reset_index()
)

# 5.1 Total matches played between each pair of teams
# Need pairs from match metadata (if available). If not, derive from team_match_runs by grouping matches with 2 teams.
# Build matches -> set of teams per match
teams_per_match = team_match_runs.groupby('match_id')['batting_team'].unique().reset_index()
# only keep matches with exactly 2 teams
teams_per_match['teams_sorted'] = teams_per_match['batting_team'].apply(lambda arr: tuple(sorted(arr)) if len(arr)==2 else None)
pairs = teams_per_match.dropna(subset=['teams_sorted']).groupby('teams_sorted').size().reset_index(name='matches_played')
pairs[['team_a','team_b']] = pd.DataFrame(pairs['teams_sorted'].tolist(), index=pairs.index)
pairs = pairs[['team_a','team_b','matches_played']].sort_values(by='matches_played', ascending=False)
print("Total matches played between each pair of teams (sample):")
print(pairs.head(20).to_string(index=False))
print()

# 5.2 Avg score per team per match (overall)
avg_score_team = team_match_runs.groupby('batting_team')['runs_in_match'].mean().reset_index().rename(columns={'runs_in_match':'avg_score_per_match'})
print("Average score per team per match:")
print(avg_score_team.sort_values(by='avg_score_per_match', ascending=False).to_string(index=False))
print()

# 5.3 Highest-scoring matches (total runs > X)
# First compute total match runs (sum of both teams per match)
match_total_runs = team_match_runs.groupby('match_id')['runs_in_match'].sum().reset_index(name='match_total_runs')
X = 350  # example high scoring threshold; adjust as needed
high_scoring_matches = match_total_runs[match_total_runs['match_total_runs'] > X].sort_values(by='match_total_runs', ascending=False)
print(f"Highest-scoring matches (total runs > {X}):")
print(high_scoring_matches.head(20).to_string(index=False))
print()

# 5.4 Lowest-scoring matches
low_scoring_matches = match_total_runs.sort_values(by='match_total_runs').head(20)
print("Lowest-scoring matches (bottom 20):")
print(low_scoring_matches.to_string(index=False))
print()

# 5.5 Closest matches (run margin < 10)  requires match-level result data
if has_cols(df_ipl, ['match_id','winner','result_margin']):
    # assume result_margin is integer runs margin or similar; if result_margin includes 'wickets' etc handle carefully
    match_info = df_ipl[['match_id','winner','result_margin']].drop_duplicates(subset=['match_id']).dropna(subset=['result_margin'])
    # If result_margin is string, try to coerce numeric
    match_info['result_margin_num'] = pd.to_numeric(match_info['result_margin'], errors='coerce')
    close_matches = match_info[(match_info['result_margin_num'].notna()) & (match_info['result_margin_num'] < 10)]
    print("Closest matches (run margin < 10):")
    print(close_matches[['match_id','winner','result_margin']].to_string(index=False))
    print()
else:
    print(" Columns 'winner' and/or 'result_margin' not found or inconsistent  cannot compute closest matches with margin < 10.\n")

# 5.6 Most wickets in a single match (total wickets by both sides)
if 'is_wicket' in df_ipl.columns:
    wickets_per_match = df_ipl.groupby('match_id')['is_wicket'].sum().reset_index(name='total_wickets_in_match')
    print("Matches with most wickets (top 20):")
    print(wickets_per_match.sort_values(by='total_wickets_in_match', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Column 'is_wicket' not found  cannot compute wickets per match.\n")

# 5.7 Highest extras in a match
if 'extra_runs' in df_ipl.columns:
    extras_per_match = df_ipl.groupby('match_id')['extra_runs'].sum().reset_index(name='total_extras')
    print("Matches with highest extras (top 20):")
    print(extras_per_match.sort_values(by='total_extras', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Column 'extra_runs' not found  cannot compute extras per match.\n")

# 5.8 Head-to-head dominance (win ratio)  requires winner column and match-level unique rows
if 'winner' in df_ipl.columns:
    match_winners = df_ipl[['match_id','winner']].drop_duplicates(subset='match_id')
    # create pairs list as before to know team A vs team B per match
    # we can join winners to teams_per_match
    tpm = teams_per_match[['match_id','teams_sorted']].dropna()
    tpm = tpm.merge(match_winners, on='match_id', how='left')
    # expand teams_sorted to columns
    tpm[['team_a','team_b']] = pd.DataFrame(tpm['teams_sorted'].tolist(), index=tpm.index)
    # Count wins for team_a vs team_b
    def pair_label(row):
        a,b = row['team_a'], row['team_b']
        return tuple(sorted([a,b]))
    tpm['pair'] = tpm.apply(pair_label, axis=1)
    wins = tpm.groupby(['pair','winner']).size().reset_index(name='wins')
    # Convert pair column back to two columns
    wins[['team_x','team_y']] = pd.DataFrame(wins['pair'].tolist(), index=wins.index)
    # pivot wins
    wins_pivot = wins.pivot_table(index=['team_x','team_y'], columns='winner', values='wins', fill_value=0)
    # Compute win ratio for pair (team_x vs team_y) by counting wins from match_winners for matches where pair matches
    # Simpler: compute pair-level total matches and wins per team
    pair_matches = tpm.groupby('pair').size().reset_index(name='matches')
    pair_wins = match_winners.copy()
    # For each pair, count wins for each team
    pair_wins = tpm.groupby(['teams_sorted','winner']).size().reset_index(name='wins')
    pair_wins[['team_a','team_b']] = pd.DataFrame(pair_wins['teams_sorted'].tolist(), index=pair_wins.index)
    pair_wins_summary = pair_wins[['team_a','team_b','winner','wins']]
    print("Head-to-head win counts (sample):")
    print(pair_wins_summary.head(20).to_string(index=False))
    print()
else:
    print(" Column 'winner' not available  cannot compute head-to-head win ratios.\n")

# ---------------------------
# 6. Statistical & Derived Metrics
# ---------------------------
print("\n" + "="*80)
print("6. STATISTICAL & DERIVED METRICS")
print("="*80 + "\n")

# 6.1 Batting Impact Score = (Runs  SR) / 100
# Need runs and strike rate per player
if has_cols(df_ipl, ['batter','batsman_runs','ball']):
    pr = df_ipl.groupby('batter').agg(runs=('batsman_runs','sum'), balls=('ball','count')).reset_index()
    pr['sr'] = (pr['runs'] / pr['balls'] * 100).round(2)
    pr['batting_impact'] = ((pr['runs'] * pr['sr']) / 100).round(2)
    print("Top Batting Impact Scores (sample):")
    print(pr.sort_values(by='batting_impact', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Batter columns missing; cannot compute Batting Impact Score.\n")

# 6.2 Bowling Impact Score = (Wickets  20)  (Economy  5)
if has_cols(df_ipl, ['bowler','total_runs','ball']):
    bowl = df_ipl.groupby('bowler').agg(runs_conceded=('total_runs','sum'), balls=('ball','count')).reset_index()
    bowl['overs'] = bowl['balls']/6
    bowl['economy'] = (bowl['runs_conceded'] / bowl['overs']).replace([np.inf, -np.inf], 0).round(2)
    # wickets
    if 'is_wicket' in df_ipl.columns:
        wk = df_ipl[df_ipl['is_wicket']==1].groupby('bowler').size().reset_index(name='wickets')
    else:
        wk = df_ipl.groupby('bowler')['player_dismissed'].count().reset_index(name='wickets') if 'player_dismissed' in df_ipl.columns else pd.DataFrame(columns=['bowler','wickets'])
    bowl = bowl.merge(wk, on='bowler', how='left').fillna(0)
    bowl['bowling_impact'] = (bowl['wickets']*20 - bowl['economy']*5).round(2)
    print("Top Bowling Impact Scores (sample):")
    print(bowl.sort_values(by='bowling_impact', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Bowler columns missing; cannot compute Bowling Impact Score.\n")

# 6.3 Run contribution % = (Player runs / Team total)  100
if has_cols(df_ipl, ['batter','batsman_runs','batting_team']):
    player_runs = df_ipl.groupby(['batting_team','batter'])['batsman_runs'].sum().reset_index(name='player_runs')
    team_runs = df_ipl.groupby('batting_team')['batsman_runs'].sum().reset_index(name='team_runs')
    player_runs = player_runs.merge(team_runs, on='batting_team', how='left')
    player_runs['run_contrib_pct'] = (player_runs['player_runs'] / player_runs['team_runs'] * 100).round(2)
    print("Player run contribution % (sample top contributors):")
    print(player_runs.sort_values(by='run_contrib_pct', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Columns required for run contribution missing.\n")

# 6.4 Wicket contribution % = (Player wickets / Team wickets)  100
if has_cols(df_ipl, ['bowler','is_wicket','bowling_team']):
    bow_wk = df_ipl[df_ipl['is_wicket']==1].groupby(['bowling_team','bowler']).size().reset_index(name='player_wickets')
    team_wk = df_ipl[df_ipl['is_wicket']==1].groupby('bowling_team').size().reset_index(name='team_wickets')
    bow_wk = bow_wk.merge(team_wk, on='bowling_team', how='left')
    bow_wk['wicket_contrib_pct'] = (bow_wk['player_wickets'] / bow_wk['team_wickets'] * 100).round(2)
    print("Bowler wicket contribution % (sample):")
    print(bow_wk.sort_values(by='wicket_contrib_pct', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" Columns required for wicket contribution missing.\n")

# 6.5 Partnership Analysis = sum of runs between batter & non-striker pairs
# This requires ball-level 'batter' and 'non_striker' columns and match context
if has_cols(df_ipl, ['match_id','batter','non_striker','batsman_runs']):
    pairs = df_ipl.groupby(['match_id','batter','non_striker'])['batsman_runs'].sum().reset_index()
    pairs_sum = pairs.groupby(['batter','non_striker'])['batsman_runs'].sum().reset_index().rename(columns={'batsman_runs':'partnership_runs'})
    print("Top partnerships (batter & non-striker) overall (sample):")
    print(pairs_sum.sort_values(by='partnership_runs', ascending=False).head(20).to_string(index=False))
    print()
else:
    print(" 'non_striker' column missing  cannot compute partnership analysis.\n")

# 6.6 Dot-ball impact (percentage dots vs runs)
if has_cols(df_ipl, ['total_runs','ball']):
    total_balls = df_ipl.shape[0]
    dot_balls = df_ipl[df_ipl['total_runs']==0].shape[0]
    dot_pct = round(dot_balls / total_balls * 100, 2)
    print(f"Dot balls: {dot_balls} / {total_balls} ({dot_pct}%)")
    print()
else:
    print(" Required columns to compute dot-ball impact missing.\n")

# ---------------------------
# 7. Extras & Discipline Insights
# ---------------------------
print("\n" + "="*80)
print("7. EXTRAS & DISCIPLINE INSIGHTS")
print("="*80 + "\n")

# 7.1 Distribution of extras by type (wide, no-ball, bye, leg-bye)
# Some datasets split extras into wide_runs, noball_runs, bye_runs, legbye_runs
extras_cols = [c for c in ['wide_runs','noball_runs','bye_runs','legbye_runs','penalty_runs'] if c in df_ipl.columns]
if extras_cols:
    extras_by_type = df_ipl[extras_cols].sum().reset_index()
    extras_by_type.columns = ['extra_type','total']
    print("Distribution of extras by type:")
    print(extras_by_type.to_string(index=False))
    print()
else:
    # fallback: if only extra_runs exist but not broken down
    if 'extra_runs' in df_ipl.columns:
        print("Only aggregated 'extra_runs' available  distribution by type not present in dataset.")
        print()
    else:
        print("No extras columns found in dataset.")
        print()

# 7.2 Bowlers with highest extras conceded
if 'extra_runs' in df_ipl.columns and 'bowler' in df_ipl.columns:
    extras_by_bowler = df_ipl.groupby('bowler')['extra_runs'].sum().reset_index().sort_values(by='extra_runs', ascending=False)
    print("Bowlers conceding most extras (top 20):")
    print(extras_by_bowler.head(20).to_string(index=False))
    print()
else:
    print(" Cannot compute extras by bowler  'extra_runs' or 'bowler' missing.\n")

# 7.3 Teams with most disciplined bowling (least extras per over)
if 'extra_runs' in df_ipl.columns and 'bowling_team' in df_ipl.columns:
    team_extras = df_ipl.groupby('bowling_team')['extra_runs'].sum().reset_index()
    team_overs = df_ipl.groupby(['bowling_team','match_id'])['over'].nunique().groupby('bowling_team').sum().reset_index(name='overs_bowled')
    team_disc = team_extras.merge(team_overs, on='bowling_team', how='left')
    team_disc['extras_per_over'] = (team_disc['extra_runs'] / team_disc['overs_bowled']).round(2)
    print("Teams with the most disciplined bowling (extras per over):")
    print(team_disc.sort_values(by='extras_per_over').to_string(index=False))
    print()
else:
    print(" Cannot compute team discipline  'extra_runs' or 'bowling_team' missing.\n")

# 7.4 Impact of extras on match outcomes (if match result available)
if has_cols(df_ipl, ['match_id','winner']) and 'extra_runs' in df_ipl.columns:
    match_extras = df_ipl.groupby('match_id')['extra_runs'].sum().reset_index()
    # merge with winners (one row per match)
    match_winner = df_ipl[['match_id','winner']].drop_duplicates(subset='match_id')
    match_extras = match_extras.merge(match_winner, on='match_id', how='left')
    # sample: average extras in matches won by winners
    avg_extras_by_winner = match_extras.groupby('winner')['extra_runs'].mean().reset_index(name='avg_extras')
    print("Average extras in matches by winning team (sample):")
    print(avg_extras_by_winner.sort_values(by='avg_extras', ascending=False).to_string(index=False))
    print()
else:
    print(" Cannot analyze extras impact vs outcomes  'winner' or 'extra_runs' missing.\n")

print("\nAll sections completed. Tweak thresholds & filters inside the script as needed.")
```

```python
"""
advanced_ipl_insights.py
Advanced / Predictive Insights for IPL ball-by-ball dataset (df_ipl)

Functions:
 - predict_expected_runs_per_over
 - win_probability_by_over
 - plot_momentum_curve
 - batsman_bowler_network
 - clutch_performance
 - player_trend_over_time
 - detect_over_scoring_anomalies

Assumes df_ipl contains at least:
 - match_id, inning, over, ball, batting_team, bowling_team,
   batsman_runs, total_runs, batter (or batsman), bowler, winner (optional),
   season / date (for trends), is_wicket (optional), result_margin (optional)
"""

import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor, GradientBoostingClassifier, IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, roc_auc_score, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import statsmodels.api as sm
from collections import defaultdict
import warnings
warnings.filterwarnings("ignore")
sns.set(style="whitegrid")

# ---------------------------
# UTILITIES
# ---------------------------
def check_columns(df, cols):
    missing = [c for c in cols if c not in df.columns]
    if missing:
        raise ValueError(f"Missing required columns: {missing}")

# ---------------------------
# 1) Predict expected runs per over (regression)
# ---------------------------
def predict_expected_runs_per_over(df, features=None, model=None, test_size=0.2, random_state=42, verbose=True):
    """
    Train a regression model to predict runs in an over.
    Aggregates ball-level to over-level rows: one row per match_id, inning, over.

    Returns:
      model, X_test, y_test, preds, metrics_dict
    """
    required = ['match_id', 'inning', 'over', 'batsman_runs', 'total_runs', 'batting_team', 'bowling_team']
    check_columns(df, required)

    # Aggregate to over level
    over_df = df.groupby(['match_id','inning','over','batting_team','bowling_team']).agg(
        runs_in_over=('total_runs','sum'),
        balls_in_over=('ball','count'),
        fours=('batsman_runs', lambda x: (x==4).sum()),
        sixes=('batsman_runs', lambda x: (x==6).sum())
    ).reset_index()

    # Feature engineering
    over_df['over_number'] = over_df['over']
    # One can add team-level historical stats (rolling means). For simplicity add previous over runs (lag per match+inning)
    over_df = over_df.sort_values(['match_id','inning','over'])
    over_df['prev_over_runs'] = over_df.groupby(['match_id','inning'])['runs_in_over'].shift(1).fillna(0)
    over_df['is_powerplay'] = over_df['over'].between(1,6).astype(int)
    over_df['is_middle'] = over_df['over'].between(7,15).astype(int)
    over_df['is_death'] = over_df['over'].between(16,20).astype(int)

    # Categorical encode batting & bowling team using simple target encoding (mean runs)
    team_mean = over_df.groupby('batting_team')['runs_in_over'].mean().rename('bat_team_mean').to_dict()
    bowl_mean = over_df.groupby('bowling_team')['runs_in_over'].mean().rename('bowl_team_mean').to_dict()
    over_df['bat_team_mean'] = over_df['batting_team'].map(team_mean)
    over_df['bowl_team_mean'] = over_df['bowling_team'].map(bowl_mean)

    # Final features
    default_features = ['over_number','prev_over_runs','balls_in_over','fours','sixes',
                        'is_powerplay','is_middle','is_death','bat_team_mean','bowl_team_mean']
    features = features or default_features
    X = over_df[features].fillna(0)
    y = over_df['runs_in_over']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)

    if model is None:
        model = RandomForestRegressor(n_estimators=200, random_state=random_state, n_jobs=-1)

    model.fit(X_train, y_train)
    preds = model.predict(X_test)

    mse = mean_squared_error(y_test, preds)
    rmse = np.sqrt(mse)
    metrics = {'rmse': rmse, 'mse': mse}

    if verbose:
        print(f"[Regression] RMSE: {rmse:.3f}")

    # Plot predicted vs actual
    plt.figure(figsize=(6,6))
    plt.scatter(y_test, preds, alpha=0.4)
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
    plt.xlabel("Actual runs in over")
    plt.ylabel("Predicted runs in over")
    plt.title("Predicted vs Actual (Over-level)")
    plt.show()

    return model, X_test, y_test, preds, metrics

# ---------------------------
# 2) Win probability curve by over (if result data available)
# ---------------------------
def win_probability_by_over(df, min_matches=200, verbose=True):
    """
    Train a classifier to estimate win probability at each over state.
    Model inputs: runs_so_far, wickets_so_far, overs_left, target (if chasing) etc.
    Returns trained model and plots a win probability curve aggregated by over number.
    """
    # required columns
    check_columns(df, ['match_id','inning','over','ball','batting_team','total_runs','winner'])

    # We need match-level aggregation of state by over per side.
    # Build cumulative runs & wickets per batting side per match+inning by over
    df_sorted = df.sort_values(['match_id','inning','over','ball'])
    # cumulative by match+inning
    df_sorted['cum_runs'] = df_sorted.groupby(['match_id','inning'])['total_runs'].cumsum()
    if 'is_wicket' in df_sorted.columns:
        df_sorted['cum_wickets'] = df_sorted.groupby(['match_id','inning'])['is_wicket'].cumsum()
    else:
        df_sorted['cum_wickets'] = 0

    # sample one row per over (last ball of the over) to represent state at that over
    over_state = df_sorted.groupby(['match_id','inning','over','batting_team']).agg(
        runs_so_far=('cum_runs','max'),
        wickets_so_far=('cum_wickets','max')
    ).reset_index()

    # merge winner to the state (winner is team name)
    winners = df[['match_id','winner']].drop_duplicates('match_id')
    over_state = over_state.merge(winners, on='match_id', how='left')
    # target: did batting_team ultimately win? (1/0)
    over_state['target_win'] = (over_state['batting_team'] == over_state['winner']).astype(int)
    # feature: overs_left for that inning (assuming 20 overs)
    over_state['overs_left'] = 20 - over_state['over']
    over_state['over_number'] = over_state['over']

    # simple features
    X = over_state[['runs_so_far','wickets_so_far','overs_left','over_number']].fillna(0)
    y = over_state['target_win']

    # require some minimum data
    if len(over_state) < min_matches:
        if verbose:
            print("Warning: limited data for training win-prob model")

    # train classifier
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
    clf = GradientBoostingClassifier(n_estimators=200, random_state=42)
    clf.fit(X_train, y_train)
    probs = clf.predict_proba(X_test)[:,1]
    auc = roc_auc_score(y_test, probs)
    if verbose:
        print(f"[WinProb] AUC: {auc:.3f}")

    # Get predicted win probability by over (simulate typical states)
    # For visualization, we compute average predicted probability per over_number across rows
    over_state['pred_prob'] = clf.predict_proba(X)[:,1]
    summary = over_state.groupby('over_number')['pred_prob'].mean().reset_index()

    plt.figure(figsize=(10,5))
    plt.plot(summary['over_number'], summary['pred_prob'], marker='o')
    plt.xlabel("Over")
    plt.ylabel("Avg Win Probability")
    plt.title("Win Probability Curve by Over (averaged across matches)")
    plt.ylim(0,1)
    plt.grid(True)
    plt.show()

    return clf, summary

# ---------------------------
# 3) Momentum visualization (cumulative runs vs overs)
# ---------------------------
def plot_momentum_curve(df, match_id=None, batting_team=None, figsize=(10,5)):
    """
    Plot cumulative runs vs overs for a given match or for aggregated team-level average per over.
    If match_id provided, plots that match; otherwise plots average cumulative runs per over for batting_team.
    """
    check_columns(df, ['match_id','over','ball','total_runs','batting_team'])

    if match_id is not None:
        match_df = df[df['match_id']==match_id].sort_values(['inning','over','ball'])
        match_df['cum_runs'] = match_df.groupby(['inning'])['total_runs'].cumsum()
        # sample last ball of each over to plot
        sample = match_df.groupby(['inning','over'])['cum_runs'].max().reset_index()
        plt.figure(figsize=figsize)
        for inn in sample['inning'].unique():
            s = sample[sample['inning']==inn]
            plt.plot(s['over'], s['cum_runs'], label=f'Inning {inn}')
        plt.xlabel("Over")
        plt.ylabel("Cumulative Runs")
        plt.title(f"Momentum Curve - Match {match_id}")
        plt.legend()
        plt.show()
        return sample
    else:
        if batting_team is None:
            raise ValueError("Provide batting_team for aggregated momentum (or match_id for match-level).")
        team = df[df['batting_team']==batting_team].sort_values(['match_id','inning','over','ball'])
        team['cum_runs'] = team.groupby(['match_id','inning'])['total_runs'].cumsum()
        sample = team.groupby(['match_id','inning','over'])['cum_runs'].max().reset_index()
        avg = sample.groupby('over')['cum_runs'].mean().reset_index()
        plt.figure(figsize=figsize)
        plt.plot(avg['over'], avg['cum_runs'], marker='o')
        plt.xlabel("Over")
        plt.ylabel("Avg Cumulative Runs")
        plt.title(f"Average Momentum Curve  {batting_team}")
        plt.show()
        return avg

# ---------------------------
# 4) Batsman vs Bowler matchup network graph
# ---------------------------
def batsman_bowler_network(df, top_n_batsmen=50, top_n_bowlers=50, min_runs=10, layout='spring'):
    """
    Build and plot a bipartite network where edges are total runs scored by batter off a bowler.
    Node sizes are proportional to runs (batter) or wickets (bowler).
    """
    check_columns(df, ['batter','bowler','batsman_runs'])

    # aggregate runs per batter-bowler
    pair = df.groupby(['batter','bowler'])['batsman_runs'].sum().reset_index()
    pair = pair[pair['batsman_runs'] >= min_runs]

    # limit to top N batters/bowlers by total
    top_batters = df.groupby('batter')['batsman_runs'].sum().nlargest(top_n_batsmen).index
    top_bowlers = df.groupby('bowler')['batsman_runs'].sum().nlargest(top_n_bowlers).index
    pair = pair[pair['batter'].isin(top_batters) & pair['bowler'].isin(top_n_bowlers)]

    G = nx.Graph()
    # add nodes
    batter_totals = df.groupby('batter')['batsman_runs'].sum().to_dict()
    bowler_totals = df.groupby('bowler')['batsman_runs'].sum().to_dict()

    for b in set(pair['batter'].unique()):
        G.add_node(f"B_{b}", label=b, bipartite=0, size=batter_totals.get(b, 10))

    for bw in set(pair['bowler'].unique()):
        G.add_node(f"R_{bw}", label=bw, bipartite=1, size=bowler_totals.get(bw,10))

    # add edges with weight = runs
    for _, r in pair.iterrows():
        G.add_edge(f"B_{r['batter']}", f"R_{r['bowler']}", weight=r['batsman_runs'])

    # layout and draw
    plt.figure(figsize=(12,12))
    if layout == 'spring':
        pos = nx.spring_layout(G, k=0.5, iterations=50)
    else:
        pos = nx.kamada_kawai_layout(G)

    # draw nodes and edges with sizes
    node_sizes = [G.nodes[n]['size']*0.5 for n in G.nodes()]
    edge_widths = [max(0.5, G[u][v]['weight'] / 50) for u,v in G.edges()]
    nx.draw_networkx_edges(G, pos, alpha=0.4, width=edge_widths)
    nx.draw_networkx_nodes(G, pos, node_size=node_sizes, node_color=['#1f78b4' if n.startswith('B_') else '#33a02c' for n in G.nodes()])
    labels = {n: G.nodes[n]['label'] for n in G.nodes() if G.nodes[n]['size'] > 200}
    nx.draw_networkx_labels(G, pos, labels, font_size=8)
    plt.title("Batsman vs Bowler Matchup Network (edge weight = runs scored)")
    plt.axis('off')
    plt.show()

    return G

# ---------------------------
# 5) Clutch performance  players who perform best under pressure
# ---------------------------
def clutch_performance(df, pressure_def):
    """
    Evaluate player performance under pressure.
    pressure_def: function(row_over_state_df) -> boolean indicating whether that ball/over is 'pressure' state
    Example simple pressure_def: lambda row: (row['over'] >= 16) & (row['match_run_rate'] > threshold)
    We'll compute batsman metrics in pressure vs non-pressure and compare.
    """
    # need ball level columns: match_id, over, ball, batter, batsman_runs, total_runs
    check_columns(df, ['match_id','over','ball','batter','batsman_runs'])

    tmp = df.copy()
    # compute small features per ball / over to decide pressure
    # Example pressure = death overs and chasing within 10 runs of target - this requires target info; we craft a generic pressure flag through user supplied function
    tmp['is_pressure'] = tmp.apply(pressure_def, axis=1).astype(int)

    # batting aggregate under pressure
    pressure_stats = tmp.groupby(['batter','is_pressure']).agg(
        runs=('batsman_runs','sum'),
        balls=('ball','count'),
        avg=('batsman_runs', lambda x: x.sum() / (x.ne(0).sum() if x.ne(0).sum() else 1))
    ).reset_index()

    # pivot pressure vs non-pressure
    pivot = pressure_stats.pivot(index='batter', columns='is_pressure', values=['runs','balls']).fillna(0)
    # flatten
    pivot.columns = ['_'.join([str(c) for c in col]).strip() for col in pivot.columns.values]
    pivot = pivot.reset_index().rename(columns={
        'runs_1':'runs_pressure','balls_1':'balls_pressure',
        'runs_0':'runs_nonpressure','balls_0':'balls_nonpressure'
    }).fillna(0)

    # compute rates
    pivot['sr_pressure'] = (pivot['runs_pressure'] / pivot['balls_pressure'] * 100).replace([np.inf, -np.inf], 0).round(2)
    pivot['sr_nonpressure'] = (pivot['runs_nonpressure'] / pivot['balls_nonpressure'] * 100).replace([np.inf, -np.inf],0).round(2)
    pivot['sr_delta'] = pivot['sr_pressure'] - pivot['sr_nonpressure']

    # rank by positive delta and minimum balls in pressure to ensure significance
    pivot['min_pressure_balls'] = 10
    result = pivot[pivot['balls_pressure'] >= pivot['min_pressure_balls']].sort_values(by='sr_delta', ascending=False)
    return result[['batter','runs_pressure','balls_pressure','sr_pressure','sr_nonpressure','sr_delta']].head(50)

# ---------------------------
# 6) Performance trend over time (career growth)
# ---------------------------
def player_trend_over_time(df, player_name, role='batter', time_col='season'):
    """
    Plot rolling average runs (batter) or wickets (bowler) per season/time_col.
    time_col may be 'season' or 'date' (ensure date is datetime).
    """
    if role == 'batter':
        check_columns(df, ['batter','batsman_runs', time_col, 'match_id'])
        tmp = df.groupby([time_col,'match_id','batter'])['batsman_runs'].sum().reset_index()
        player = tmp[tmp['batter'] == player_name].groupby(time_col)['batsman_runs'].mean().reset_index(name='avg_runs_per_match')
        plt.figure(figsize=(8,4))
        plt.plot(player[time_col], player['avg_runs_per_match'], marker='o')
        plt.title(f"{player_name}  Avg Runs per Match by {time_col}")
        plt.xlabel(time_col)
        plt.ylabel('Avg Runs')
        plt.show()
        return player
    else:
        check_columns(df, ['bowler','is_wicket', time_col, 'match_id'])
        tmp = df[df['is_wicket']==1].groupby([time_col,'match_id','bowler']).size().reset_index(name='wickets_per_match')
        player = tmp[tmp['bowler'] == player_name].groupby(time_col)['wickets_per_match'].mean().reset_index()
        plt.figure(figsize=(8,4))
        plt.plot(player[time_col], player['wickets_per_match'], marker='o')
        plt.title(f"{player_name}  Avg Wickets per Match by {time_col}")
        plt.xlabel(time_col)
        plt.ylabel('Avg Wickets')
        plt.show()
        return player

# ---------------------------
# 7) Anomaly detection (unexpected high/low scoring over)
# ---------------------------
def detect_over_scoring_anomalies(df, contamination=0.01, features=None):
    """
    Use IsolationForest on over-level aggregated features to flag anomalous overs (very high or low scoring)
    Returns DataFrame of unusual overs with anomaly score.
    """
    check_columns(df, ['match_id','inning','over','total_runs','batsman_runs','ball'])

    over_df = df.groupby(['match_id','inning','over']).agg(
        runs_in_over=('total_runs','sum'),
        balls_in_over=('ball','count'),
        fours=('batsman_runs', lambda x: (x==4).sum()),
        sixes=('batsman_runs', lambda x: (x==6).sum())
    ).reset_index()

    default_features = ['runs_in_over','balls_in_over','fours','sixes']
    features = features or default_features
    X = over_df[features].fillna(0)

    iso = IsolationForest(contamination=contamination, random_state=42)
    iso.fit(X)
    over_df['anomaly_score'] = iso.decision_function(X)
    over_df['anomaly'] = iso.predict(X)  # -1 = anomaly, 1 = normal

    anomalies = over_df[over_df['anomaly'] == -1].sort_values('anomaly_score')
    print(f"Detected {len(anomalies)} anomalous overs (contamination={contamination})")
    return anomalies

# ---------------------------
# Example usage (commented)
# ---------------------------
if __name__ == "__main__":
    # load your df_ipl first (not included here)
    # Example calls (uncomment & adapt):
    #
    # model, X_test, y_test, preds, metrics = predict_expected_runs_per_over(df_ipl)
    # clf, win_summary = win_probability_by_over(df_ipl)
    # avg_cum = plot_momentum_curve(df_ipl, batting_team='Mumbai Indians')
    # G = batsman_bowler_network(df_ipl)
    # anomalies = detect_over_scoring_anomalies(df_ipl, contamination=0.01)
    #
    # Example clutch def:
    # def simple_pressure(row):
    #     return (row['over'] >= 16) and (row['match_id'] % 2 == 0)  # placeholder; customize
    # clutch = clutch_performance(df_ipl, simple_pressure)
    #
    # trend = player_trend_over_time(df_ipl, 'V Kohli', role='batter', time_col='season')
    #
    pass
```

```python
# ============================================================
#  Simple IPL Analytics Dashboard (Streamlit)
# ============================================================

import streamlit as st
import pandas as pd
import plotly.express as px

# ------------------------------------------------------------
# Page Setup
# ------------------------------------------------------------
st.set_page_config(page_title="IPL Analytics Dashboard", layout="wide")
st.title(" IPL Comprehensive Analytics Dashboard")

# ------------------------------------------------------------
# Upload / Load Data
# ------------------------------------------------------------
st.sidebar.header(" Data Input")
uploaded_file = st.sidebar.file_uploader("Upload IPL CSV file", type=["csv"], accept_multiple_files=False)

if uploaded_file is not None:
    df = pd.read_csv(uploaded_file)
else:
    st.info("Upload `ipl_data.csv` to get started.")
    st.stop()

# ------------------------------------------------------------
# Sidebar Filters
# ------------------------------------------------------------
teams = sorted(df['batting_team'].unique())
players = sorted(df['batter'].unique())

selected_team = st.sidebar.selectbox("Select Team", teams)
selected_player = st.sidebar.selectbox("Select Player", players)

# ------------------------------------------------------------
# Tabs for Sections
# ------------------------------------------------------------
tabs = st.tabs([" Overview", " Team Insights", " Player Insights", " PDF Report"])

# ------------------------------------------------------------
#  Overview
# ------------------------------------------------------------
with tabs[0]:
    st.subheader("General / Overview Insights")
    total_matches = df['match_id'].nunique()
    total_teams = df['batting_team'].nunique()
    total_runs = df['total_runs'].sum()
    total_wickets = df['is_wicket'].sum()
    total_balls = len(df)
    total_overs = total_balls / 6
    run_rate = total_runs / total_overs

    c1, c2, c3 = st.columns(3)
    c1.metric("Total Matches", total_matches)
    c2.metric("Total Teams", total_teams)
    c3.metric("Total Runs", f"{total_runs:,}")

    c4, c5, c6 = st.columns(3)
    c4.metric("Total Wickets", f"{total_wickets:,}")
    c5.metric("Total Overs", f"{total_overs:,.0f}")
    c6.metric("Run Rate", f"{run_rate:.2f}")

    st.divider()
    st.write("### Runs Distribution by Team")
    team_runs = df.groupby('batting_team')['total_runs'].sum().sort_values(ascending=False)
    fig = px.bar(team_runs, x=team_runs.index, y=team_runs.values, title="Total Runs by Team", color=team_runs.values, color_continuous_scale="sunset")
    st.plotly_chart(fig, use_container_width=True)

# ------------------------------------------------------------
#  Team Insights
# ------------------------------------------------------------
with tabs[1]:
    st.subheader(f"Team Analysis  {selected_team}")
    team_df = df[df['batting_team'] == selected_team]

    total_team_runs = team_df['total_runs'].sum()
    total_team_wkts = team_df['is_wicket'].sum()
    overs_team = len(team_df) / 6
    rr_team = total_team_runs / overs_team

    st.write(f"**Total Runs:** {total_team_runs:,} | **Wickets:** {total_team_wkts:,} | **Run Rate:** {rr_team:.2f}")

    phase_df = team_df.copy()
    phase_df['phase'] = pd.cut(phase_df['over'], bins=[0,6,15,20], labels=['Powerplay','Middle','Death'])
    phase_summary = phase_df.groupby('phase')['total_runs'].mean()

    fig = px.bar(phase_summary, x=phase_summary.index, y=phase_summary.values, title="Average Runs per Over by Phase", color=phase_summary.values)
    st.plotly_chart(fig, use_container_width=True)

# ------------------------------------------------------------
#  Player Insights
# ------------------------------------------------------------
with tabs[2]:
    st.subheader(f"Player Analysis  {selected_player}")
    player_df = df[df['batter'] == selected_player]

    if len(player_df) > 0:
        total_runs_player = player_df['batsman_runs'].sum()
        balls_faced = len(player_df)
        fours = len(player_df[player_df['batsman_runs'] == 4])
        sixes = len(player_df[player_df['batsman_runs'] == 6])
        dismissals = player_df['is_wicket'].sum()
        avg = total_runs_player / dismissals if dismissals > 0 else total_runs_player
        sr = (total_runs_player / balls_faced) * 100

        st.write(f"**Runs:** {total_runs_player:,} | **Average:** {avg:.2f} | **Strike Rate:** {sr:.2f} | **4s:** {fours} | **6s:** {sixes}")

        st.divider()
        st.write("### Runs by Over")
        over_summary = player_df.groupby('over')['batsman_runs'].sum()
        fig = px.line(over_summary, x=over_summary.index, y=over_summary.values, markers=True, title=f"{selected_player} - Runs by Over")
        st.plotly_chart(fig, use_container_width=True)
    else:
        st.warning("No data available for this player in the dataset.")

# ------------------------------------------------------------
#  PDF Report
# ------------------------------------------------------------
with tabs[3]:
    st.subheader(" IPL Insights PDF Report")
    try:
        with open("IPL_Comprehensive_Match_Insights.pdf", "rb") as pdf_file:
            st.download_button(" Download PDF Report", data=pdf_file, file_name="ipl_report.pdf", mime="application/pdf")
        st.write("Preview of your report:")
        st.pdf("IPL_Comprehensive_Match_Insights.pdf")
    except FileNotFoundError:
        st.error("No `ipl_report.pdf` found in your folder. Please add it and refresh.")

# ------------------------------------------------------------
# End of App
# ------------------------------------------------------------
st.caption(" IPL Analytics Dashboard | Built with Streamlit")
```
